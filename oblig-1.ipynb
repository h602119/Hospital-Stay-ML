```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt



```

```python
training_data = pd.read_csv('./training_data.csv')
test_data = pd.read_csv('./test_data.csv')

#Configure pandas such that all columns are displayed.
pd.set_option('display.max_columns', None)
```

The training data has genders represented as Female and Male, but since this can easier be represented as numbers, I change them to 0.0 and 1.0, where F is 0.0 and M is 1.0.
<br><br>
The faculty id is represented as chars, this has to be converted to numbers where A is 0.0, B is 1.0 and so on.

```python
#Swapping genders = F to 0.0 and Gender = M to 1.0
training_data['gender'] = training_data['gender'].replace({'F': 0.0, 'M': 1.0})
```

```python
#Sorting the unique faculty ids
faculty_ids = sorted(training_data['facid'].unique())

#The current numb to
current_num = 0.0
for char in faculty_ids:
  training_data['facid'] = training_data['facid'].replace({char : current_num})
  current_num += 1.0
```

```python
#The visit date 'vdate', is not important for the model
#Hence we will remove it.
training_data = training_data.drop('vdate', axis=1)
```

```python
display(training_data)
```

All data is not collected on all patients, which means some of them have null (NAN) values, and this sparks a dilemma.
<br>
* Should I drop the rows where a nan is present, this might be ok if the dataset is large.
* Set the values to an average or median value, this might be smart if I have smaller datasets, or have little to lose when setting a median value.

As we can see below 1.21% of the data is un-complete.

```python
#Select all rows where a nan is present
rows_with_nan = training_data.isna().any(axis=1).sum()

# Print the amount of rows with NaN values
print(f"Rows with NAN present: {rows_with_nan}")
print(f"This is equivalent of {round(rows_with_nan * 100 / 70000,2)}% of the training data")
```

Based on the low percentage of un-complete data, I have decided to remove the data from the training data. This is because I do not wish to influence the outcome in any way, by setting a median or average value.


```python
#Remove all rows where a NAN value is found
training_data_cleaned = training_data.dropna()
```

```python
from sklearn.model_selection import train_test_split
#The data is already split into training and test data

#Split the training data into variables and result.
X = training_data_cleaned.drop('lengthofstay', axis=1)
y = training_data_cleaned['lengthofstay']
X_train, X_val, y_train, y_val = train_test_split(X, y)

#The test data has no 'length of stay' so I will have to predict this.
test_X = test_data.copy()

```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Instantiate and fit the model
regressor = RandomForestRegressor(n_estimators=128, random_state=20)
regressor.fit(X_train, y_train)
```

```python
#Using our trained model to predict visit length on the test data
predicted_test_y = regressor.predict(test_X)

display(predicted_test_y)
```

```python
#Visualizing the percentage of 'length of stay' in the training and test data.

#Training data
training_dictionary = {}

for number in y_train:
  if number in training_dictionary:
    training_dictionary[number] += 1
  else:
    training_dictionary[number] = 1

for entry in training_dictionary:
  training_dictionary[entry] = round(training_dictionary[entry] * 100 / 70000,3)

#Sorting the dictionary
training_dictionary = {k: v for k, v in sorted(training_dictionary.items())}

print(training_dictionary)



#Test data
test_dictionary = {}

for number in predicted_test_y:
  number = round(number)
  if number in test_dictionary:
    test_dictionary[number] += 1
  else:
    test_dictionary[number] = 1

for entry in test_dictionary:
  test_dictionary[entry] = round(test_dictionary[entry] * 100 / 30000,3)

#Sorting the dictionary
test_dictionary = {k: v for k,v in sorted(test_dictionary.items())}

print(test_dictionary)

plot_train_x = training_dictionary.values()
plot_train_y = training_dictionary.keys()
plt.plot(plot_train_x,plot_train_y,'ro')
plt.show()

plot_test_x = test_dictionary.values()
plot_test_y = test_dictionary.keys()
plt.plot(plot_test_x,plot_test_y,'bo')
plt.show()
```

Im going to use the R^2 Score to measure the goodness of the fit of this model.



```python
# Calculate the R-squared score on the validation data
r2_score_regressor = regressor.score(X_val, y_val)


print(f"the score is {r2_score_regressor}")
```
